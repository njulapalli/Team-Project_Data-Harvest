```python
import matplotlib as notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Data used from below URL.
# https://www.ers.usda.gov/data-products/vegetables-and-pulses-data/vegetables-and-pulses-yearbook-tables/#Supply%20and%20Utilization:%20Fresh%20Market

# File to Load 
Veggie_Supply_file = "VegetablesandpulsesCSV/Vegetables_and_Pulses_Yearbook_Supply_and_Utilization_Data_2018.csv"

# Read Purchasing File and store into Pandas data frame
Veggie_Supply_df = pd.read_csv(Veggie_Supply_file)
Veggie_Supply_df.dtypes
```

```python
Veggie_Supply_df['PublishValue'] = Veggie_Supply_df["PublishValue"].str.replace(",","").astype(float)
Veggie_Supply_df.dtypes
```

```python
Veggie_Supply_df.head(10)
```

# 2017 price of vegetables

```python
veg_fresh_price1_2017_df = Veggie_Supply_df[ Veggie_Supply_df.Year == 2017]
veg_fresh_price2_2017_df = veg_fresh_price1_2017_df[ veg_fresh_price1_2017_df.EndUse == "Fresh"]
veg_fresh_price_2017_df = veg_fresh_price2_2017_df[ veg_fresh_price2_2017_df.Item == "Current Dollars"]

veg_fresh_price_2017_df =  veg_fresh_price_2017_df.sort_values(['PublishValue'], ascending=False)
veg_fresh_price_2017_df
```

```python
x_axis = veg_fresh_price_2017_df['Commodity']
y_axis = veg_fresh_price_2017_df['PublishValue']

#plt.figure(figsize=(10,4))
plt.figure(figsize=(10,5))
plt.bar(x_axis, y_axis, facecolor='green', alpha=0.75, align="center")
plt.title("Fresh Vegetable Prices in 2017")
plt.ylabel("Average Price($/cwt)")
plt.xticks(rotation='vertical')
plt.show()
#plt.xlabel()

```

# 2017 per capita availability of vegetables

```python
veg_fresh_capita1_2017_df = Veggie_Supply_df[ Veggie_Supply_df.Year == 2017]
veg_fresh_capita2_2017_df = veg_fresh_capita1_2017_df[ veg_fresh_capita1_2017_df.EndUse == "Fresh"]
veg_fresh_capita3_2017_df = veg_fresh_capita2_2017_df[ veg_fresh_capita2_2017_df.Commodity != "Total vegetables"]
veg_fresh_capita_2017_df = veg_fresh_capita3_2017_df[ veg_fresh_capita3_2017_df.Item == "Per Capita Availability"]

veg_fresh_capita_2017_df = veg_fresh_capita_2017_df.sort_values(['PublishValue'], ascending=False)
veg_fresh_capita_2017_df
```

```python
top_5_veg = veg_fresh_capita_2017_df.head(5).Commodity
all_veg = veg_fresh_capita_2017_df.Commodity

x_axis = veg_fresh_capita_2017_df['Commodity']
y_axis = veg_fresh_capita_2017_df['PublishValue']



#f, ax=plt.subplots(figsize=(18,5))
plt.figure(figsize=(10,5))
plt.bar(x_axis, y_axis, facecolor='green', alpha=0.75, align="center")
plt.title("Fresh Vegetable Per Capita Availability in 2017")
plt.ylabel("Pounds")
plt.xticks(rotation='vertical')


plt.show()
#plt.xlabel()


```

```python
top_5_veg
```

```python
for ind, vegetable in top_5_veg.iteritems():
    veg_fresh1_2017_df = Veggie_Supply_df[ Veggie_Supply_df.Year > 1990]
    veg_fresh2_2017_df = veg_fresh1_2017_df[ veg_fresh1_2017_df.EndUse == "Fresh"]
    veg_fresh3_2017_df = veg_fresh2_2017_df[ veg_fresh2_2017_df.Commodity == vegetable]
    veg_fresh_2017_df = veg_fresh3_2017_df[ veg_fresh3_2017_df.Item == "Per Capita Availability"]
    
    x_axis = veg_fresh_2017_df['Year']
    y_axis = veg_fresh_2017_df['PublishValue']
    plt.plot(x_axis, y_axis, label = vegetable, marker='o')

#Legends
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="Commodity", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
#ax.legend(fontsize = 24)
# Set your x and y limits
plt.xlim(1990, 2018)
plt.ylim(0, 60)
plt.title("Fresh Vegetable Per Capita Availability Trendline from 1990 to 2017")
plt.ylabel("Pounds")
plt.grid()
fig = plt.gcf()
fig.set_size_inches(10, 7)

plt.show()
```

# Price graph for all Vegetables

```python
for ind, vegetable in top_5_veg.iteritems():
    veg_price1_df = Veggie_Supply_df[ Veggie_Supply_df.Year > 2000]
    veg_price2_df = veg_price1_df[ veg_price1_df.EndUse == "Fresh"]
    veg_price3_df = veg_price2_df[ veg_price2_df.Commodity == vegetable]
    veg_price_df = veg_price3_df[ veg_price3_df.Item == "Current Dollars"]
    
    x_axis = veg_price_df['Year']
    y_axis = veg_price_df['PublishValue']
    plt.plot(x_axis, y_axis, label = vegetable, marker='o')

#Legends
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="Commodity", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
#ax.legend(fontsize = 24)
# Set your x and y limits
plt.xlim(2000, 2018)
plt.ylim(0, 80)
plt.title("Fresh Vegetable Price Trendline from 2000 to 2017")
plt.ylabel("Price (Dollars per hundred pounds)")
fig = plt.gcf()
fig.set_size_inches(10, 7)
plt.grid()
plt.show()
```

# Revenue earned for each Vegetables in 2017

Data Analysis for Vegetables

```python
veg_fresh_capita1_df = Veggie_Supply_df[ Veggie_Supply_df.Year > 1990]
veg_fresh_capita2_df = veg_fresh_capita1_df[ veg_fresh_capita1_df.EndUse == "Fresh"]
veg_fresh_capita3_df = veg_fresh_capita2_df[ veg_fresh_capita2_df.Commodity != "Total vegetables"]
veg_fresh_capita_df = veg_fresh_capita3_df[ veg_fresh_capita3_df.Item == "Per Capita Availability"]
veg_fresh_price_df = veg_fresh_capita3_df[ veg_fresh_capita3_df.Item == "Current Dollars"]

veg_fresh_capita_df.head(10)

#veg_fresh_capita_2017_df = veg_fresh_capita_2017_df.sort_values(['PublishValue'], ascending=False)
#veg_fresh_capita_2017_df
```

```python
veg_fresh_price_df = veg_fresh_price_df.rename(columns={'Category': 'PriceCategory', 'Item': 'PriceItem', 
                                   'Unit': 'PriceUnit', 'PublishValue': 'Price'})

veg_fresh_price_df.head(10)
```

```python
# Combine the data into a single dataset
veg_fresh_revenue_df = pd.merge(veg_fresh_capita_df, veg_fresh_price_df, how="left", 
                                on=["Decade", "Year", "Commodity","EndUse"])
veg_fresh_revenue_df ['RevenuePerCapita'] = (veg_fresh_revenue_df['PublishValue'] * veg_fresh_revenue_df['Price']) / 100

veg_fresh_revenue_df = veg_fresh_revenue_df.round({'RevenuePerCapita' : 2})
veg_fresh_revenue_df.head(10)
```

```python
for ind, vegetable in top_5_veg.iteritems():
    graph_revenue_df = veg_fresh_revenue_df[ veg_fresh_revenue_df.Commodity == vegetable]
    x_axis = graph_revenue_df['Year']
    y_axis = graph_revenue_df['RevenuePerCapita']
    plt.plot(x_axis, y_axis, label = vegetable, marker='o')

#Legends
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="Commodity", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
#ax.legend(fontsize = 24)
# Set your x and y limits
plt.xlim(1990, 2018)
plt.ylim(0, 12)
plt.title("Revenue Per Capita Trendline of all Vegetables from 1990 to 2017")
plt.ylabel("Revenue ($/capita)")
fig = plt.gcf()
fig.set_size_inches(10, 7)
plt.grid()
plt.show()

```
